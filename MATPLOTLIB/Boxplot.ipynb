```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,vert=False)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,widths=0.2)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=["python"])

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=["python"],patch_artist=True)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=["python"],showmeans=True)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=["python"],showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"))

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=["python"],showmeans=True,sym="g+",boxprops=dict(color="r"),capprops=dict(color="r"),whiskerprops=dict(color="r"),flierprops=dict(markeredgecolor="y"))

plt.show()
```
