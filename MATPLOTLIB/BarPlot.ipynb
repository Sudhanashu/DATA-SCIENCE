```python
#BAR Plot
#BASIC BAR PLOT
import matplotlib.pyplot as plt

x=["Python","C","C++","Java"]
y=[85,70,60,82]
plt.bar(x,y)
plt.show()
```

```python
#LABELS AND TITLE

import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]

plt.bar(x, y)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity Score')
plt.title('Popularity of Programming Languages')
plt.show()
```

```python
#MULTIPLE BAR PLOT
import matplotlib.pyplot as plt

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20,30,40,50]

c=["y","b","m","g"]

plt.bar(x, y,width=0.3,color="r",label="popularity")
plt.bar(x, z,width=0.3,color="y",label="popularity1")
plt.legend()
plt.xlabel('Programming Languages',fontsize=15)
plt.ylabel('Popularity Score',fontsize=15)
plt.title('Popularity of Programming Languages',fontsize=15)
plt.show()
```

```python
#MULTIPLE BAR PLOT
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20,30,40,50]

width=0.2

p=np.arange(len(x))
p1=[j+width for j in p]

plt.bar(p, y,width,color="r",label="popularity")
plt.bar(p1, z,width,color="y",label="popularity1")
plt.legend()

plt.xlabel('Programming Languages',fontsize=15)
plt.ylabel('Popularity Score',fontsize=15)
plt.title('Popularity of Programming Languages',fontsize=15)
plt.show()
```

```python
#MULTIPLE BAR PLOT
import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20,30,40,50]

width=0.2

p=np.arange(len(x))
p1=[j+width for j in p]

plt.bar(p, y,width,color="r",label="popularity")
plt.bar(p1, z,width,color="y",label="popularity1")


plt.xlabel('Programming Languages',fontsize=15)
plt.ylabel('Popularity Score',fontsize=15)
plt.title('Popularity of Programming Languages',fontsize=15)

plt.xticks(p+width/2,x)
plt.legend()
plt.show()
```

```python
#MULTIPLE BAR PLOT
#ROTATION

import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20,30,40,50]

width=0.2

p=np.arange(len(x))
p1=[j+width for j in p]

plt.bar(p, y,width,color="r",label="popularity")
plt.bar(p1, z,width,color="y",label="popularity1")


plt.xlabel('Programming Languages',fontsize=15)
plt.ylabel('Popularity Score',fontsize=15)
plt.title('Popularity of Programming Languages',fontsize=15)

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```

```python
#MULTIPLE BAR PLOT
#HORIZONTAL BAR GRAPH

import matplotlib.pyplot as plt
import numpy as np

x = ["Python", "C", "C++", "Java"]
y = [85, 70, 60, 82]
z = [20,30,40,50]

width=0.2

p=np.arange(len(x))
p1=[j+width for j in p]

plt.barh(p, y,width,color="r",label="popularity")
plt.barh(p1, z,width,color="y",label="popularity1")


plt.xlabel('Programming Languages',fontsize=15)
plt.ylabel('Popularity Score',fontsize=15)
plt.title('Popularity of Programming Languages',fontsize=15)

plt.xticks(p+width/2,x,rotation=20)
plt.legend()
plt.show()
```
