```python
#SCATTER PLOT
#BASIC SCATTER PLOT

import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.show()
```

```python
#TITLE AND FONTSIZE

import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7]
y=[2,3,1,4,5,3,6]
plt.scatter(x,y)
plt.title("Scatter Plot",fontsize=20)
plt.show()
```

```python
#LABEL

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#COLOR

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
plt.scatter(day,no,color="r")
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#MULTIPLE COLOR

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
plt.scatter(day,no,c=colors)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#MULTIPLE SIZES

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[50,100,250,40,150,75,180]
plt.scatter(day,no,c=colors,s=sizes)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#TRANSPARENCY OR ALPHA PARAMETER

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[50,100,250,40,150,75,180]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.7)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#MARKER

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[50,100,250,40,150,75,180]
plt.scatter(day,no,c=colors,s=sizes,alpha=0.7,marker="*")
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#EDGE COLOR

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[50,100,250,40,150,75,180]
plt.scatter(day,no,c=colors,s=sizes,marker="*",edgecolor="g")
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#LINE WIDTH

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=["r","y","g","b","r","g","r"]
sizes=[50,100,250,40,150,75,180]
plt.scatter(day,no,c=colors,s=sizes,marker="*",edgecolor="g",linewidth=2)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#CMAP OR COLOR MAP

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,30,29,56,20,30]
sizes=[50,100,250,40,150,75,180]
plt.scatter(day,no,c=colors,s=sizes,cmap="viridis")
plt.colorbar()
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#CMAP OR COLOR MAP

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
colors=[10,49,30,29,56,20,30]
sizes=[50,100,250,40,150,75,180]
plt.scatter(day,no,c=colors,s=sizes,cmap="BrBG")
t = plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```

```python
#CMAP OR COLOR MAP

import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
no=[2,3,1,4,5,3,6]
no2=[3,2,4,5,1,6,2]
colors=[10,49,30,29,56,20,30]
sizes=[50,100,250,40,150,75,180]
plt.scatter(day,no,c=colors,s=sizes,cmap="BrBG")
plt.scatter(day,no2,c=colors,s=sizes,cmap="BrBG")
t = plt.colorbar()
t.set_label("ColorBar",fontsize=15)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day",fontsize=20)
plt.ylabel("no",fontsize=20)
plt.show()
```
