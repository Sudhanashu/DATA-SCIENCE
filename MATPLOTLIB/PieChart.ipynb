```python
#PIE PLOT

import matplotlib.pyplot as plt
x = [10,20,30,40]
y=["c","c++","java","python"]

plt.pie(x,labels=y)

plt.show()
```

```python
#LABELS AND ITS LOCATION

import matplotlib.pyplot as plt
x = [10,20,30,40]
y=["c","c++","java","python"]
ex=[0.4,0.0,0.0,0.0]
c=["r","b","g","y"]

plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,startangle=0,textprops={"fontsize": 15},counterclock=False,wedgeprops={"linewidth":5,"edgecolor":"m"},rotatelabels=True)

plt.title("Wscube Tech")
plt.legend(loc=2)
plt.show()
```

```python
#MULTIPLE PIE PLOT

import matplotlib.pyplot as plt
x = [10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","java","python"]

c=["r","b","g","y"]

plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=0.5,colors=c)

plt.show()
```

```python
#CIRCLE PIE PLOT

import matplotlib.pyplot as plt
x = [10,20,30,40]
x1=[40,30,20,10]
y=["c","c++","java","python"]

c=["r","b","g","y"]

plt.pie(x,labels=y,radius=1.5)
cr = plt.Circle(xy=(0,0),radius=1,facecolor="w")
plt.gca().add_artist(cr)

plt.show()
```
