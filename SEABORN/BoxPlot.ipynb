```python
#BOX PLOT

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
var = sns.load_dataset("tips")
var
```

```python
#GRID and basic box plot

sns.set(style="whitegrid")

sns.boxplot(x="day", y="total_bill", data=var)
plt.show()
```

```python
#HUE PARAMETER

sns.set(style="whitegrid")

sns.boxplot(x="day", y="total_bill", data=var,hue="time")
plt.show()
```

```python
#HUE PARAMETER

#sns.set(style="whitegrid")

sns.boxplot(x="day", y="total_bill", data=var,hue="sex")
plt.show()
```

```python
#COLOR PARAMETER

sns.set(style="whitegrid")

sns.boxplot(x="day", y="total_bill", data=var,hue="sex",color="g")
plt.show()
```

```python
#ORDER PARAMETER

sns.set(style="whitegrid")

sns.boxplot(x="day", y="total_bill", data=var,hue="sex",color="g",order=["Fri","Sun","Thur","Sat"])
plt.show()
```

```python
#MEAN PARAMETER

sns.set(style="whitegrid")

sns.boxplot(x="day", y="total_bill", data=var,hue="sex",color="g",order=["Fri","Sun","Thur","Sat"],showmeans=True)
plt.show()
```

```python
#MEANPROPS PARAMETER

sns.set(style="whitegrid")

sns.boxplot(x="day", y="total_bill", data=var,hue="sex",color="g",order=["Fri","Sun","Thur","Sat"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"})
plt.show()
```

```python
#LINEWIDTH PARAMETER

sns.set(style="whitegrid")

sns.boxplot(x="day", y="total_bill", data=var,hue="sex",color="g",order=["Fri","Sun","Thur","Sat"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3)
plt.show()
```

```python
#PALETTE PARAMETER

sns.set(style="whitegrid")

sns.boxplot(x="day", y="total_bill", data=var,hue="sex",color="g",order=["Fri","Sun","Thur","Sat"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette='plasma')
plt.show()
```

```python
#HORIZONTAL PARAMETER

sns.set(style="whitegrid")

sns.boxplot(y="day", x="total_bill", data=var,hue="sex",color="g",order=["Fri","Sun","Thur","Sat"],showmeans=True,meanprops={"marker":"+","markeredgecolor":"b"},linewidth=3,palette='plasma')
plt.show()
```

```python
#Boxplot with single parameter

sns.set(style="whitegrid")

sns.boxplot(x=var["total_bill"])

plt.show()
```
