```python
#STRIP PLOT

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.stripplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
#HUE PARAMETER

sns.stripplot(x="day",y="total_bill",data=var,hue="sex")
plt.show()
```

```python
#PALETTE PARAMETER

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r")
plt.show()
```

```python
#LINEWIDTH PARAMETER

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1.5)
plt.show()
```

```python
#EDGECOLOR PARAMETER

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1.5,edgecolor="r")
plt.show()
```

```python
#JITTER PARAMETER

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1.5,edgecolor="r",jitter=5)
plt.show()
```

```python
#SIZE PARAMETER

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1.5,edgecolor="r",size=5)
plt.show()
```

```python
#MARKER PARAMETER

m={"Male":"o","Female":"<"}

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1.5,edgecolor="r",size=5,marker="<")
plt.show()
```

```python
#ALPHA PARAMETER

m={"Male":"o","Female":"<"}

sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="rocket_r",linewidth=1.5,edgecolor="r",size=5,marker="<",alpha=0.5)
plt.show()
```

```python
#STRIP PLOT WITH SINGLE PARAMETER

sns.stripplot(y=var["total_bill"])
```
