```python
#VIOLIN PLOT PLAYS A SIMILAR ROLE AS A BOX AND WHISKER PLOT. IT SHOWS THE DISTRIBUTION OF QUANTATIVE DATA ACROSS SEVERAL LEVELS OF ONE(OR MORE) CATEGORICAL VARIABLES SUCH THAT THOSE DISTRIBUTIONS CAN BE COMPARED

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var)
plt.show()
```

```python
#HUE PARAMETERS

sns.violinplot(x="day",y="total_bill",data=var,hue="time")
plt.show()
```

```python
#LINEWIDTH PARAMETERS

sns.violinplot(x="day",y="total_bill",data=var,linewidth=3)
plt.show()
```

```python
#PALETTE PARAMETERS

sns.violinplot(x="day",y="total_bill",data=var,linewidth=3,palette="Dark2")
plt.show()
```

```python
#ORDER PARAMETERS

sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"])
plt.show()
```

```python
#COLOR PARAMETERS

sns.violinplot(x="day",y="total_bill",data=var,linewidth=3,color="r")
plt.show()
```

```python
#SPLIT PARAMETERS

sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True)
plt.show()
```

```python
#SLACE PARAMETERS
#IN SCALE PARAMETERS WE CAN GIVE "COUNT","AREA","WIDTH" ETC AS A PARAMETER

sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="count")
plt.show()
```

```python
#SLACE PARAMETERS
#IN SCALE PARAMETERS WE CAN GIVE "COUNT","AREA","WIDTH" ETC AS A PARAMETER

sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=True,scale="width")
plt.show()
```

```python
#HORIZONTAL PARAMETER

sns.violinplot(y="day",x="total_bill",data=var,hue="sex")
plt.show()
```

```python
#INNER PARAMETR
#IN INNER PARAMTER YOU CAN USE "QUART","QUARTILE","POINT","STICK","BOX" ETC AS A PARAMETER

sns.violinplot(x="time",y="total_bill",data=var,order=["Dinner","Lunch"],inner="quart")
plt.show()
```

```python
#SINGLE VIOLIN PLOT

sns.violinplot(x=var["total_bill"])
plt.show()
```
