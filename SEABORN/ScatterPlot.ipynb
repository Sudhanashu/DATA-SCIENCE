```python
#SCATTER PLOT

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var)

plt.show()
```

```python
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex")

plt.show()
```

```python
#STYLE PARAMETERS

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex")

plt.show()
```

```python
#Size PARAMETERS

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(10,40))

plt.show()
```

```python
#COLOR/PALLETE PARAMETERS

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(10,40),palette='gist_ncar')

plt.show()
```

```python
#ALPHA PARAMETERS

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(10,40),palette='gist_ncar',alpha=0.75)

plt.show()
```

```python
#MARKERS PARAMETERS

m={"Male":"*","Female":"o"}

sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var,hue="sex",style="sex",size="sex",sizes=(10,40),palette='gist_ncar',alpha=0.75,markers=m)

plt.show()
```
