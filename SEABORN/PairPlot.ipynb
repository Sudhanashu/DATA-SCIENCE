```python
#PAIR PLOT

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.pairplot(var)
plt.show()
```

```python
#HUE PARAMETERS

sns.pairplot(var,hue="sex")
plt.show()
```

```python
sns.pairplot(var,vars=["total_bill","tip"],hue="sex")
plt.show()
```

```python
#HUE_ORDER PARAMETERS

sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Feamle","Male"])
plt.show()
```

```python
#PALETTE PARAMETERS

sns.pairplot(var,vars=["total_bill","tip"],hue="sex",hue_order=["Feamle","Male"],palette="BuGn")
plt.show()
```

```python
sns.pairplot(var,x_vars=["total_bill","tip"],hue="sex",hue_order=["Feamle","Male"],palette="BuGn")
plt.show()
```
