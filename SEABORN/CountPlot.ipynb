```python
#WHAT IS DIFFERENCE BETWEEN COUNTPLOT AND BARPLOT
#-> COUNTPLOT PLOTS THE THE COUNT OF THE NUMBER OF THE RECORDS BY CATEGORY
#->BARPLOT PLOTS  A VALUE OR METRIC FOR EACH CATEGORY (BY DEFAULT, BARPLOT PLOTS THE MEAN OF A VARIABLE, BY CATEGORY)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.countplot(x="sex",data = var)
plt.show()
```

```python
sns.barplot(x="sex",y="size",data=var)
plt.show()
```

```python
#HUE PARAMETER

sns.countplot(x="sex",hue="smoker",data=var)
plt.show()
```

```python
#HORIZONTAL PARAMETER

sns.countplot(y="sex",hue="smoker",data=var)
plt.show()
```

```python
#PALETTE PARAMETER

sns.countplot(x="sex",hue="smoker",data=var,palette="bwr")
plt.show()
```

```python
#COLOR PARAMETER AND PALETTE PARAMETER ALTHOUGH BOTH ARE SAME

sns.countplot(x="sex",hue="smoker",data=var,color="y")
plt.show()
```

```python
#SATURATION PARAMETER

sns.countplot(x="sex",hue="smoker",data=var,color="r",saturation=0.5)
plt.show()
```
