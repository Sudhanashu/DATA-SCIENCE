```python
#FACTOR PLOT 
#NOTE:- FACTOR PLOT IS NOW RENAMED AS CATPLOT

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.catplot(x="size",y="tip",data=var)
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,hue="sex")
plt.show()
```

```python
sns.catplot(x="size",y="tip",data=var,hue="sex",kind="box")
plt.show()
```
